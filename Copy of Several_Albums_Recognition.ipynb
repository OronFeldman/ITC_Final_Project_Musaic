```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imutils
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('IMG-20240331-WA0005.jpg')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(img_gray, (5, 5), 0)
edges_Canny = cv2.Canny(img, 150, 250)
edges_Canny_equalized = cv2.equalizeHist(edges_Canny)
```

```python
contours, _ = cv2.findContours(edges_Canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

contour_img = img.copy()
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)
cv2_imshow(contour_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
